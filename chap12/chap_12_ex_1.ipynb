# Programmation dynamique et suite de Fibonacci

Rappel : F(1)=1; F(2)=1; F(3)=F(2)+F(1)=2, F(4)=F(3)+F(2)=3 ... F(n)=F(n-1)+F(n-2)

## Avec une fonction récursive classique

```python
def fibonacci(n):
    if n<=2:
        return 1
    return fibonacci(n-1)+fibonacci(n-2)
```

```python
fibonacci(4)
```

Pour compter les appels récursifs à la fonction $fibonacci$, on met en place un compteur :

```python
compteur=0
def fibonacci(n):
    global compteur
    compteur+=1
    if n<=2:
        return 1
    return fibonacci(n-1)+fibonacci(n-2)
```

```python
fibonacci(4)
```

```python
compteur
```

Ce résultat est vérifiable si on visualise les différents appels sur un arbre :

```python
#Pour remplir le tableau
n=[3,4,5,6,7,8,9,10]
appels=[]
for i in n:
    compteur=0
    fibonacci(i)
    appels.append(compteur)
print(n)
print(appels)
```

```python
ymax=max(appels)
import matplotlib.pyplot as plt
plt.scatter(n, appels, c='red')
plt.xlabel("n")
plt.ylabel("nombre d'appels")
plt.ylim(0,ymax)
plt.show()
```

## En utilisant la programmation dynamique

On utilise un dictionnaire pour mémoriser les valeurs déjà calculées de la suite de Fibonacci.

```python
compteur=0

def fibonacci_recursive(n,d) :
    global compteur
    compteur+=1
    if n<=2 :
        return 1
    if n in d : #la valeur a déjà été calculée !
        return d[n] 
    valeur_retournee = fibonacci_recursive(n-1, d) + fibonacci_recursive(n-2, d)
    d[n] = valeur_retournee
    return valeur_retournee

def fibonacci(n) :
    dico_fibo = {}
    return fibonacci_recursive(n, dico_fibo)
```

```python
#Pour remplir le tableau
n=[3,4,5,6,7,8,9,10]
appels=[]
for i in n:
    compteur=0
    fibonacci(i)
    appels.append(compteur)
print(n)
print(appels)
```

Il est possible de vérifier ce nombre d'appels récursifs sur l'exemple de Fibonacci(5). Les appels qui seront effectués seront les suivants : F(5), F(4), F(3), F(2), F(1), F(2), F(3) soit 7 au total. L'arbre est parcouru de manière préfixe. On gagne du temps sur le parcours du sous-arbre droit car F(3) est connue.

```python
import matplotlib.pyplot as plt
plt.scatter(n, appels, c='green')
plt.xlabel("n")
plt.ylabel("nombre d'appels")
plt.ylim(0,ymax)
plt.show()
```
